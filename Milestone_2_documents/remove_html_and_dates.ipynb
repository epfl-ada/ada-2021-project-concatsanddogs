```python
def remove_html_tags(df,column='quotation'):
    return df[~df[column].str.contains('|'.join(('.*?>', '<.*?', '<.*?>')))]
```

```python
remove_html_tags(df,column='quotation')
```

```python
# detects if a row has a (or has a part of) a html tag in quote and then removes it
df[~df['quotation'].str.contains('|'.join(('.*?>', '<.*?', '<.*?>')))] 
```

```python
from dateutil.parser import parse

def check_if_date(string, fuzzy = True):
    try: 
        parse(string, fuzzy=fuzzy)
        return True
    except ValueError:
        return False
```

```python
df[df['quotation'].apply(lambda x : check_if_date(x))] # this tells what quotations are dates ( mistakes on 1 and two)
```
